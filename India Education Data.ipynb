## Part 3: Visualization using Matplot

Matplotlib can be used in Python scripts, the Python and IPython shell, the jupyter notebook, web application servers, and four graphical user interface toolkits.

```python
# lets load the education data set using pandas
eduindia = pd.read_csv("C:\\Users\\pnkum\\Desktop\\Project\\IndiaEducationData\\final_data.csv") 

# Lets us see the first 10 entries of the file.
eduindia.head(10)
```

```python
# For this example we will see rural india, you can change rural to urban or total to get different graph with the same code
india1 = eduindia.loc[eduindia['Type'].isin(['Rural'])]
```

### What is a line chart?

A line chart or line graph is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. It is similar to a scatter plot except that the measurement points are ordered (typically by their x-axis value) and joined with straight line segments.

Source: https://en.wikipedia.org/wiki/Line_chart 

Example: We will compare total female population with the female with a graduate degree.

```python
#We set figure size
fig,ax = plt.subplots(figsize=(10,10))

# We plot state(in code) vs Total female population. 
plt.plot(india1.Code, india1.TFemales, 'red')

# We plot state(in code) vs Graduate female population. 
plt.plot(india1.Code, india1.GFemales, 'blue')

# This code is used to mention the title for your visualization
plt.title('Total Females vs Female Graduate population')

# This code is sued to label your x axis
plt.xlabel('State Code')

# This code is sued to label your x axis
plt.ylabel('population')

# Final use the code to show/dispaly the graph
plt.show()
```

```python
# The other way of using the line graph is by showing the count for each category
# In below graph we see that we are gouping by part and comparing sum of Graduate males and graduate female
indishow=india1.groupby(by='part').sum()[['GMales','GFemales']]

# We can also use single line command like below to mention title, kind of graph and figure size
indishow.plot(kind='line',figsize=(12,8),grid=True,title='Graduate Male vs Graduate Female')
```

Using subplot we can compare all the values of female education like the Total female, total literate female, the literate female without education, female with the Non-Technical diploma, Technical diploma, and graduate female.

```python
# We  can also plot mulitiple graph in the same graph by plotting each graph and mentioning their position
# This is the size of the graph
fig,ax = plt.subplots(figsize=(10,10))
# This is mention the position of the graph in relation to other fraph
plt.subplot(3,2,1)
#This is to mention what to plot on the graph and color of the graph
plt.plot(india1.Code, india1.TFemales, 'blue')
plt.title('Total Female')
plt.subplot(3,2,3)
plt.plot(india1.Code, india1.NTDFemales,'green')
plt.title('Total Female with Non Technical Diploma')
plt.subplot(3,2,2)
plt.plot(india1.Code, india1.LFemales, 'red')
plt.title('Total Literate Female')
plt.subplot(3,2,4)
plt.plot(india1.Code, india1.LWEFemales, 'brown')
plt.title('Total Female who are literate without education')
plt.subplot(3,2,5)
plt.plot(india1.Code, india1.TDFemales, 'black')
plt.title('Total Female with Technical Diploma')
plt.subplot(3,2,6)
plt.plot(india1.Code, india1.GFemales)
plt.title('Total Female With Graduation')
# We used this command to put all the plot in same graph
plt.tight_layout()
#command used to show the graph
plt.show()
```

#### Observation
1. We see that in spite of large female population the total graduate are really low
2. We see that there is very low graduate female across the state in India, irrespective of the total female population.
3. This graph shows the gap between the region and also between males and female population within the region.
To classify the result better, we will classify region wise using a bar chart.

### What is a barplot?
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

Soruce: https://en.wikipedia.org/wiki/Bar_chart

Let us check the states with region by plotting a bar graph and pie chart

```python
fig,ax = plt.subplots(figsize=(8,8))
india1['part'].value_counts(sort=False).plot(kind='bar',color = 'blue')
plt.title('Number of state per region',fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.ylabel('Number of State',fontsize=15)
```

For the above region, lets compare Graduate males vs graduate female. This compares two value using sum function and compare them in the same graph as shown below

```python
indicomp=india1.groupby(by='part').sum()[['TFemales','GFemales']]
indicomp.plot(kind='bar',figsize=(12,8),grid=True,title='Comparing total Total females vs graduate females for every region')
```

We see that same graph above can be viewed horizontally by changing one code from 'bar' to 'barh'. Sometimes this gives a better visualization.

```python
ipm2=india1.groupby(by='part').sum()[['TFemales','GFemales']]
ipm2.plot(kind='barh',figsize=(12,8),grid=True,title='Comparing total Total females vs graduate females for every region')
```

#### Observation

We see in every region the female graduate are far lower than total female population.

### What is scatter plot?

A scatter plot (also called a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram)[3] is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are color-coded, one additional variable can be displayed. ## Part 4: Visualization using Seaborn

Seaborn is a Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics.

For more information, you can visit: https://seaborn.pydata.org/

We will try to plot same scatter plot we did in the last example, but you can see that in seaborn we can give colors to the different states and add legends. Which looks more pleasing than the previous scatter plot.

```python
# We'll use seaborn's FacetGrid to color the scatterplot by AreaName and also add the legend
# We will use \ command to move the code to next line so it is readable
sns.FacetGrid(india1, hue="AreaName", size=8) \
   .map(plt.scatter, "TFemales", "GFemales") \
   .add_legend()
```

We will use seaborn joint plot shows bivariate scatterplots and univariate histograms with regression line density estimation in the same figure

```python
sns.jointplot(x="TFemales", y="GFemales", data=india1, size=8, kind='reg', color='blue')
```

We will use seaborn joint plot shows bivariate scatterplots and univariate histograms with Kernel density estimation in the same figure

```python
sns.jointplot(x="TFemales", y="GFemales", data=india1, size=6, kind='kde', color='blue', space=0)
```

We will use seaborn joint plot shows bivariate scatterplots and univariate histograms with hex graph in the same figure

```python
sns.jointplot(x="TFemales", y="GFemales", data=india1, size=6, kind='hex', color='#800000')
```

We can use boxplot which is a standardized way of displaying the distribution of data based on the five-number summary: minimum, first quartile, median, third quartile, and maximum.

```python
sns.boxplot(x="part", y="GFemales", data=india1)
```

We try to strip the plot and use jitter to see all the point for each species and save it in axes as ax

```python
ax = sns.boxplot(x="part", y="GFemales", data=india1)
ax = sns.stripplot(x="part", y="GFemales", data=india1, jitter=True, edgecolor="gray")
```

We use swarm plot to plot the data

```python
sns.swarmplot(x="part", y="GFemales", data=india1)
```

We use violin plot to plot the data

```python
sns.violinplot(x="part", y="GFemales", data=india1, size=10)
```

Seaborn plot is useful for looking at univariate relations is the kdeplot, which creates and visualizes a kernel density estimate of the underlying feature

```python
sns.FacetGrid(india1, hue="part", size=6) \
   .map(sns.kdeplot, "GFemales") \
   .add_legend()
```
